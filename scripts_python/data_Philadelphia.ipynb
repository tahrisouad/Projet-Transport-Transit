philadelphia by_route

```python
import pandas as pd
# Chargement de Philadelphie
path_philly_route = "../data_raw/Average_Daily_Ridership_By_Route - City of Philadelphia.csv"
df_philly_route = pd.read_csv(path_philly_route)
print(df_philly_route.head())
```

```python
#EDA: Philadelphie By Route
import seaborn as sns
import matplotlib.pyplot as plt

# la Taille de data (Lines & Columns)
shape = df_philly_route.shape
print(f"--- la Taille de Dataframe ---")
print(f"Nombre de lignes: {shape[0]}")
print(f"Nombre de colonnes: {shape[1]}")

# Qualité Data (Nulls & Doublons)
print("\n--- Qualité de Data ---")
print(f"Valeurs nulles par colonne:\n{df_philly_route.isnull().sum()}")
print(f"\nNombre de doublons: {df_philly_route.duplicated().sum()}")

# Description de la DataFrame

print("Description de la DataFrame Route:  \n",df_philly_route.describe())

# les infos de la DataFrame

print("Info sur la DataFrame Route: \n", df_philly_route.info())
```

```python
# Exportation du DataFrame final au format CSV(BY_Route)
df_philly_route.to_csv("../Data/philadelphie_route.csv", index=False)
print("Fichier exporté avec succès !")
```
